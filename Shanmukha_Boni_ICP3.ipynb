```python
import numpy as np
```

```python
r=np.random.randint(1,21,size=15)
print(r)

r=r.reshape(3,5)
print(r)

print(r.shape)

for i in range(r.shape[0]):
    j = r[i]
    j[j == max(r[i])]=0

print(r)
```

```python
r2 = np.array([[3,-2],[1,0]])
e_val,e_vec = np.linalg.eig(r2)
print(e_val)
print(e_vec)
```

```python
r3 = np.array([[0,1,2],[3,4,5]])
sum=0
for i in range(r3.shape[0]):
    for j in range(r3.shape[1]):
      if i==j:
        sum+=r3[i][j]

print(sum)
```
